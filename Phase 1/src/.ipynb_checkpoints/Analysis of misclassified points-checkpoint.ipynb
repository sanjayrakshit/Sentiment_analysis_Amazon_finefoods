```python
import pandas as pd
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
sns.set()
```

### Loading the data: 
***

```python
with open('../data/misclassified.pkl','rb') as fp:
    data = pickle.load(fp)
fp.close()
```

```python
data.columns
```

```python
plt.figure(figsize=(10,5))
# Histogram of the 'pred' column
plt.hist(x=data.pred,normed=True,rwidth=0.6,bins=[0,1,2])
plt.xticks([0+.5,1+.5],['Misclassified as Negative','Misclassified as Positive'])
plt.title('Misclassified Prediction',fontdict={'fontsize':30})
plt.show()
```

So around 80% of the misclassified points are actually **positive** and the rest is **negative**

```python
predicted_1 = data[data.pred == 1]
```

```python
predicted_0 = data[data.pred == 0]
```

```python
pred_X_0 = predicted_0.X.tolist()
pred_X_1 = predicted_1.X.tolist()
```

```python
pred_X_0 = ' '.join(pred_X_0)
pred_X_1 = ' '.join(pred_X_1)
```

```python
plt.figure(figsize=(20,10),facecolor='w')
'''Creating a wordcloud of reviews that were misclassified as NEGATIVE'''
wc = WordCloud(background_color='black',width=900,height=500).generate(pred_X_0)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)
plt.title('WC of *Reviews that were misclassified as negative*',fontdict={'fontsize':25})

plt.show()
```
